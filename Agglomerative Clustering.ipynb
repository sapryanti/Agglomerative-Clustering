```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn import datasets
```

```python
from sklearn.datasets import make_blobs
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:,:2]
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X[:,0],X[:,1],'ro')
```

<b>Langkah Ke 2<b>

```python
clus = AgglomerativeClustering(n_clusters=3, linkage="average", affinity='euclidean')
```

```python
clus.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clus.labels_,marker='x')
plt.show()
```

<b>langkah 3<b>

```python
clus1 = AgglomerativeClustering(n_clusters=3, linkage="complete", affinity='euclidean')
```

```python
clus1.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clus1.labels_,marker='x')
plt.show()
```
